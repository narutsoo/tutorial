### Regression Feature Selection
(Numerical input, Numerical output)

* Feature selection is performed used Pearson's correlation Coefficient via f_regression

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
# Generate dataset
X, y = make_regression(n_samples=1000, n_features=100, n_informative=10)
```

```python
# Define Feature Selection
fs = SelectKBest(score_func=f_regression, k=10)
# Apply Feature Selection
X_selected = fs.fit_transform(X, y)
```

```python
print (X_selected.shape)
```

### Classification Feature Selection
(Numerical input, Categorical output)

```python
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
# Generate dataset
X, y = make_classification(n_samples=100, n_features=10, n_informative=2)
```

```python
# Define Feature Selection
fs = SelectKBest(score_func=f_classif, k=2)
# Apply Feature Selection
X_selected = fs.fit_transform(X, y)
```

```python
print (X_selected.shape)
```
